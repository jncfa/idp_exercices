```python
# necessary imports

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
 
# define matplotlib style
plt.style.use("ggplot")

plt.rcParams.update({'font.size': 14})
```

```python
f = pd.read_csv("u_data", sep=';')
print(f)
print(f.columns)
```

```python
fig = plt.figure(figsize=[11,7])#(figsize=[8,8])
axis = fig.add_subplot()
axis.plot(f[f['e']!=1]['e'],f[f['e']!=1][' u'], 'o', markersize=5)
axis.plot(f[f['e']==1]['e'],f[f['e']==1][' u'], 'v', markersize=5)
#axis.set_title("Distance: {}m".format(int(distance/100)))
axis.legend(loc=(0.025, 0.5))
axis.set_xscale('log')
axis.set_yscale('log')
axis.set_ylabel(r'Coeficiente de atenuaçao mássico $\mu_m$ $(cm^2/g)$')
axis.set_xlabel(r'Energia do fotão $h\nu$ (MeV)')

#twinaxis.legend(loc=(0.3, 0.73))
#twinaxis.xaxis.get_major_formatter().labelOnlyBase = False
fig.savefig(r"images/u_plot.pdf", facecolor='white')
```
