```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

plt.style.use("ggplot")
FILENAME_HEADER = "filename"
LAMBDA_HEADER = "lambda"
path = "/home/jncfa/idp_exercices/class2/tpc/"

filename = path + 'values.log'
```

```python
plt.figure()

values = None
with open(filename, 'r') as file:
    values = [tuple(float(x.strip()) for x in l.split(",")) for l in file.readlines()] 

ax = plt.axes(projection='3d')
ax.scatter3D([a[0] for a in values], [a[1] for a in values], [a[2] for a in values])

plt.show()
```
