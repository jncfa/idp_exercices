```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
FILENAME_HEADER = "filename"
LAMBDA_HEADER = "lambda"
path = "/home/jncfa/idp_exercices/class1/tpc/"

filedata = [{FILENAME_HEADER: path+"values1a-1.log", LAMBDA_HEADER: 1}, {FILENAME_HEADER: path+"values1a-2.log", LAMBDA_HEADER: 2}]
```

```python
for currentFile in filedata:
    with open(currentFile[FILENAME_HEADER], 'r') as file:
        values = [float(l.strip()) for l in file.readlines()]
        print(sum(values)/len(values))
        plt.hist(values, bins=100, label=r"$\lambda={}$".format(currentFile[LAMBDA_HEADER]))


plt.ylabel('Count of numbers')
plt.xlabel('Distribution Value')
plt.setp(plt.legend().get_texts(), color='k') # for some reason the text is showing up white??
plt.show()
```
